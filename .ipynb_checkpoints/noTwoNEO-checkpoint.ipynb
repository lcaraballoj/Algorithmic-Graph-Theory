```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import hypernetx as hnx
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
from cycler import cycler
```

# 4 Hyperedges

```python
scenes = {"e0": [3, 4], "e1": [2, 4], "e2": [1, 3], "e3": [1, 2]}

H = hnx.Hypergraph(scenes)
print("Adjacency Matrix: \n", H.adjacency_matrix().toarray())

print("\n Incidence Matrix: \n", H.incidence_matrix().toarray())

# ax.axis('off')
pos = {
    1: (-4,4),
    2: (4,4),
    3: (-4,-4),
    4: (4,-4)
}

fig,ax = plt.subplots(1,1,figsize=(8,8))

hnx.drawing.draw(H,
    ax=ax,
    pos=pos,
    node_labels_kwargs={
        'fontsize': 48
    },
    with_edge_labels=False,
    edges_kwargs={
        'linewidths': 6
    }
)
```
